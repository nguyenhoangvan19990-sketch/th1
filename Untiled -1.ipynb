```python
a=float(input("Chiều rộng của khu đất HCN là :"))
b=float(input("Chiều dài của khu đất HCN là :"))
dien_tich_khu_dat_HCN= a*b
pi=3.14
R=a/2
dien_tich_khu_vui_choi= pi*R*R
S= dien_tich_khu_dat_HCN - dien_tich_khu_vui_choi
print("Diện tích còn lại của khu đất là :",S,round(S,2))
```

```python
a=float(input("Cạnh T1 của Tam giác là :"))
b=float(input("Cạnh T2 của Tam giác là :"))
c=float(input("Cạnh T3 của Tam giác là :"))
if a+b>c and a+c>b and b+c>a:
    p=(a+b+c)/2
    S=(p*(p-a)*(p-b)*(p-c))**0.5
    print("Diện tích tam giác là :",S,round(S,1))
else:
    print("Ba cạnh a,b,c không phải là cạnh của một tam giác")
```
